```python
# RSA
## Graphical Results
def plot_mtx(mtx, labels, title):
    pl.figure()
    pl.imshow(mtx, interpolation='nearest')
    pl.xticks(range(len(mtx)), labels, rotation=-45)
    pl.yticks(range(len(mtx)), labels)
    pl.title(title)
    pl.clim((0, 2))
    pl.colorbar()

roi_datatsets = [ds[:, ds.fa.IPL != 0],
                 ds[:, ds.fa.aSTS != 0],
                 ds[:, ds.fa.mSTS != 0],
                 ds[:, ds.fa.pSTS != 0]]
roi_names = ["IPL","aSTS","mSTS","pSTS"]

for roi_ds, roi_name in zip(roi_datatsets, roi_names):
    mtgs = mvpa.mean_group_sample(['person'])
    mtds = mtgs(roi_ds)
    dsm = rsa.PDist(square=True)
    res = dsm(mtds)

    plot_mtx(res, mtds.sa.person, '{}-ROI pattern correlation distances'.format(roi_name))
    pl.show()
```

```python
# Normal Classifier Analysis
clf = mvpa.kNN(k=1, dfx=mvpa.one_minus_correlation, voting='majority')

roi_datatsets = [ds[:, ds.fa.IPL != 0],
                 ds[:, ds.fa.aSTS != 0],
                 ds[:, ds.fa.mSTS != 0],
                 ds[:, ds.fa.pSTS != 0]]
roi_names = ["IPL","aSTS","mSTS","pSTS"]

# noch suboptimal durch zu viele eingeschlossene Voxel
# Ansatz aber so in Ordnung?

for roi_ds, roi_name in zip(roi_datatsets, roi_names):
    clf = mvpa.kNN(k=1, dfx=mvpa.one_minus_correlation, voting='majority')
    cvte = mvpa.CrossValidation(clf, mvpa.NFoldPartitioner(cvtype=0.85, count=10, 
                                                           selection_strategy="random",
                                                           attr="participant"),
                                errorfx=lambda p, t: np.mean(p == t),
                                enable_ca=['stats'])
    results = cvte(roi_ds)
    #print cvte.ca.stats.as_string(description=True)
    #print cvte.ca.stats.matrix

print cvte.ca.stats.as_string(description=True)
print cvte.ca.stats.matrix
```
