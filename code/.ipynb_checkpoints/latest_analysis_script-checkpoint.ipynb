```python
### RSA (between + within)
def roi_wise_dsms(rois, sub_ds, result_dict, split_num, design, num_samples_dict, **kwargs):
    participant = kwargs.get('participant', None)
    
    for roi in rois:
        # reduce to roi voxels
        roi_ds = sub_ds[:, {roi: [1]}]
        
        # zscore samples
        mvpa.zscore(roi_ds)
        
        if design == "within":
            roi_ds = roi_ds[{"participant": [participant]}, :]
        
        # Get number of included samples
        num_samples = roi_ds.shape[0]
        
        # Mean for each target
        mtgs = mvpa.mean_group_sample(['targets'])
        mtds = mtgs(roi_ds)

        # compute DSMs
        dsm = rsa.PDist(square=True) 
        res = dsm(mtds)

        # Append to result list
        if design == "between":
            key = "Split {} - {}".format(split_num, roi)
            result_dict.update({key: res})
        else:
            key = "Participant {} - Split {} - {}".format(participant,
                                                          split_num, roi)
            result_dict.update({key: res})
        
        voxel_info_key = "{} - Voxel".format(roi)
        num_samples_dict.update({key: num_samples})
        result_dict.update({voxel_info_key: roi_ds.shape[1]})
        
    return result_dict, num_samples_dict


def rsa_dsm_com(res_list, design, measure, **kwargs):
    participants = kwargs.get('participants', None)
    
    # generate mean DSM
    if design == "between":
        res_list = [res[1].samples for res in res_list]
        
        if measure == "mean":
            avg_dsm = np.mean(np.array(res_list), axis=0)
        else:
            avg_dsm = np.std(np.array(res_list), axis=0)
            
    if design == "within":
        dsm_list = []
        
        # compute DSM for each Part individually
        for participant in participants:
            key = "Participant {}".format(participant)
            part_res_list = [res[1].samples for res in res_list if key in res[0]]
            
            if measure == "mean":
                part_avg_dsm = np.mean(np.array(part_res_list), axis=0)
            else:
                part_avg_dsm = np.std(np.array(part_res_list), axis=0)
                
            dsm_list.append(part_avg_dsm)
        
        # build mean DSM over Part DSMs
        avg_dsm = np.mean(np.array(dsm_list), axis=0)
            
    return avg_dsm


def rsa_func(ds, target_list, design, **kwargs):
    inc_rois = kwargs.get('inc_rois', None)
    measure = kwargs.get('measure', "mean")
    
    result_dict = {}
    
    # only include samples of desired targets
    ds = ds[{"targets": target_list}, :]
    
    # get ROIs from ds
    if inc_rois is None:
        rois = [roi for roi in ds.fa if "ROI" in roi]
    else:
        rois = [roi for roi in ds.fa if "ROI" in roi and roi in inc_rois]
    
    # define balancer based on design
    if design == "between":
        limit = {'participant': np.unique(ds.sa.participant)}
    else: 
        limit = None
    
    balancer = mvpa.Balancer(amount="equal", attr='targets', count=10,
                             limit=limit, apply_selection=False, 
                             include_offlimit=False, space='balanced_set', 
                             rng=None)
    
    # get all partitions of balancer
    bc_sel_list = list(b.sa.balanced_set for b in balancer.generate(ds))
    
    # Store number of included samples
    num_samples_dict = {}
    
    for split_num, sel in enumerate(bc_sel_list):
        # add and apply balancer selection
        ds.sa["selection"] = sel
        sub_ds = ds[{"selection": [True]}, :]
        
        if design == "between":
            dsms = roi_wise_dsms(rois, sub_ds, result_dict,
                                 split_num, design, num_samples_dict)
            result_dict.update(dsms[0])
        else:
            for participant in np.unique(sub_ds.sa.participant):
                dsms = roi_wise_dsms(rois, sub_ds, result_dict, split_num, design, 
                                     num_samples_dict, participant=participant)
                result_dict.update(dsms[0])
                
        num_samples_dict.update(dsms[1])
    
    # Same amount of samples for each roi, print 1 example
    samples_info = np.sum([item[1] 
                           for item in num_samples_dict.items() 
                           if rois[0] in item[0]])
    print "Number of included samples: {}".format(samples_info)
    
    for roi in rois:
        res_list = []
        for item in result_dict.items():
            if roi in item[0] and "Voxel" not in item[0]:
                res_list.append(item)
            if roi in item[0] and "Voxel" in item[0]:
                voxel_info = item[1]

        if design == "between":
            avg_dsm = rsa_dsm_com(res_list, design, measure)
        else:
            participants = np.unique(ds.sa.participant)
            avg_dsm = rsa_dsm_com(res_list, design, measure, participants=participants)
        
        # Result Visualization
        print "Number of included voxels: {}".format(voxel_info)
        uf.rsa_result_plot(avg_dsm, np.unique(sub_ds.sa.targets),
                           '{}-ROI pattern correlation distances'.format(roi.split("-")[1]))
        pl.show()
    
    return result_dict

targets = ['FORREST', 'MRS. GUMP', 'FORREST (V.O.)', 'LT. DAN']
rois = ["ROI - V1", "ROI - lSTS", "ROI - laSTS", "ROI - rIPL",
        "ROI - rSTS", "ROI - raSTS", "ROI - rpsts"]

rsa_results = rsa_func(ds, targets, inc_rois=rois, measure="mean", design="between") # within


# add option to compare group of targets vs. speech
```

```python
### MVPA
def mvpa_main(ds, roi_name_list, clf_dict, design):
    result_dict = {}
    
    if "between" in design and "within" in design:
        result_between = mvpa_core(result_dict, ds, roi_name_list, clf_dict, "between")
        
        result_within = {}
        for participant in np.unique(ds.sa.participant):
            res = mvpa_core(result_dict, ds, roi_name_list, clf_dict, "within", part_num=participant)
            result_within.update(res)
            
        result_between.update(result_within)
        result = result_between
        
    elif design == "between":
        result = mvpa_core(result_dict, ds, roi_name_list, clf_dict, "between")
        
    elif design == "within":
        result_within = {}
        for participant in np.unique(ds.sa.participant):
            res = mvpa_core(result_dict, ds, roi_name_list, clf_dict, "within", part_num=participant)
            result_within.update(res)
        result = result_within
        
    return result


def mvpa_core(result_dict, ds, roi_name_list, clf_dict, mvpa_type, **kwargs):
    participant = kwargs.get('part_num')
    
    for roi in roi_name_list:
        for c in clf_dict:
            clf = clf_dict[c]
            clf.ca.enable('estimates')
            
            # Adjust parameters according to design
            if mvpa_type == "between":
                attr = "participant"
                limit = {'participant': np.unique(ds.sa.participant)}
                key = "{}: {} - {}".format("Between Subject", roi, c)
            elif mvpa_type == "within":
                attr = "chunks"
                limit = None
                key = "{}: Participant {} - {} - {}".format("Within Subject", participant, roi, c)
            
            # Set up MVPA
            cvte = mvpa.CrossValidation(clf, mvpa.NFoldPartitioner(cvtype=1, 
                                                                   attr=attr,
                                                                   space="partition"),
                                        errorfx=lambda p, t: np.mean(p == t),
                                        enable_ca=['stats']) 
            
            # Create a balanced ds
            balancer = mvpa.Balancer(amount="equal", attr='targets', count=1,
                                     limit=limit, apply_selection=True, 
                                     include_offlimit=False, space='balanced_set', 
                                     rng=None)
            balanced_ds = balancer(ds)
            
            # Compute results
            if mvpa_type == "between":
                results = cvte(balanced_ds[:, {roi: [1]}])
            elif mvpa_type == "within":
                results = cvte(balanced_ds[{'participant': [participant]}, {roi: [1]}])
            
            # Add results to dict
            result_dict[key] = cvte
    return result_dict
    
# Compute results
roi_names = ["rIPL","aSTS","rpSTS", "V1"]
clfs = {'1-Nearest-Neighbour': mvpa.kNN(k=1),
        '10-Nearest-Neighbour': mvpa.kNN(k=10)
       }

mvpa_results = mvpa_main(ds, roi_names, clfs, ["between", "within"])
```

```python
import nibabel as nib
import nilearn as nil
import numpy as np
import scipy as sp

def get_adjusted_mask(ori_mask, ref_space, **kwargs):
    overlap_mask = kwargs.get('overlap_mask', None)
    overlap = kwargs.get('overlap', "overlap")
    dilation = kwargs.get('dilation', True)
    side = kwargs.get('side', None)
    ori_mask = str(ori_mask)
    ref_space = str(ref_space)

    if overlap_mask is not None:
        overlap_mask = str(overlap_mask)
        overlap_mask_p = str(overlap_mask)

    mask = nib.load(ori_mask)
    mask_affine = mask.affine
    mask_shape = mask.shape

    ref = nib.load(ref_space)
    ref_affine = ref.affine
    ref_shape = ref.shape

    if np.array_equal(mask_affine, ref_affine) and np.array_equal(mask_shape, ref_shape):
        new_mask = mask
        new_mask_p = ori_mask
        change = 0
        
        new_mask_data = mask.get_fdata()
        
        if dilation:
            new_mask_data = sp.ndimage.binary_dilation(new_mask_data)
            
        if side is not None:
            data = new_mask_data
            mean = len(data)/2
            
            for idx, array in enumerate(data, 0):
                if side == "right" and idx > mean:
                    for idx1, mid_array in enumerate(data[idx], 0):
                        for idx2, end_array in enumerate(data[idx, idx1], 0):
                            data[idx, idx1, idx2] = False
                elif side == "left" and idx < mean:
                    for idx1, mid_array in enumerate(data[idx], 0):
                        for idx2, end_array in enumerate(data[idx, idx1], 0):
                            data[idx, idx1, idx2] = False
            new_mask_data = data
        
        if dilation or side is not None:
            new_mask_p = ori_mask[:-7]
            
            if dilation:
                new_mask_p = new_mask_p + "_dilation"
            
            if side is not None:
                new_mask_p = new_mask_p + "_{}".format(side)
                
            new_mask_p = new_mask_p + ".nii.gz"
            
        if overlap_mask is None:
            if dilation or side is not None:
                new_mask = nil.image.new_img_like(new_mask, new_mask_data, affine=ref_affine)
                nib.save(new_mask, new_mask_p)
            
    else:
        change = 1
        new_mask = nil.image.resample_img(mask, target_affine=ref_affine,
                                          target_shape=ref_shape, clip=True,
                                          interpolation="continuous")

        new_mask_data = new_mask.get_fdata()
        mask_mean = np.mean(new_mask_data)
        new_mask_data[new_mask_data >= mask_mean] = True
        new_mask_data[new_mask_data < mask_mean] = False
        new_mask_data[np.isnan(new_mask_data)] = False
        
        if dilation:
            new_mask_data = sp.ndimage.binary_dilation(new_mask_data)
        
        if side is not None:
            data = new_mask.get_fdata()
            mean = len(data)/2
            
            for idx, array in enumerate(data, 0):
                if side == "right" and idx > mean:
                    for idx1, mid_array in enumerate(data[idx], 0):
                        for idx2, end_array in enumerate(data[idx, idx1], 0):
                            data[idx, idx1, idx2] = False
                elif side == "left" and idx < mean:
                    for idx1, mid_array in enumerate(data[idx], 0):
                        for idx2, end_array in enumerate(data[idx, idx1], 0):
                            data[idx, idx1, idx2] = False
            new_mask_data = data
            
        
        new_mask = nil.image.new_img_like(new_mask, new_mask_data, affine=ref_affine)
        
        if dilation or side is not None:
            new_mask_p = "{}_adjusted".format(ori_mask[:-7])
            
            if dilation:
                new_mask_p = new_mask_p + "_dilation"
            
            if side is not None:
                new_mask_p = new_mask_p + "_{}".format(side)
                
            new_mask_p = new_mask_p + ".nii.gz"
        else:
            new_mask_p = "{}_adjusted.nii.gz".format(ori_mask[:-7])
        
    if overlap_mask is None and change == 1:
        nib.save(new_mask, new_mask_p)
        
    elif overlap_mask is not None:
        overlap_mask = nib.load(overlap_mask)
        overlap_mask_affine = overlap_mask.affine
        overlap_mask_shape = overlap_mask.shape

        if not np.array_equal(overlap_mask_affine, ref_affine) or not np.array_equal(overlap_mask_shape, ref_shape):
            overlap_mask = nil.image.resample_img(overlap_mask, target_affine=ref_affine,
                                                  target_shape=ref_shape, clip=True,
                                                  interpolation="continuous")
        mask_data = new_mask.get_fdata()
        mask_mean = np.mean(mask_data)
        mask_data[mask_data >= mask_mean] = True
        mask_data[mask_data < mask_mean] = False
        mask_data[np.isnan(mask_data)] = False

        overlap_mask_data = overlap_mask.get_fdata()
        overlap_mask_mean = np.mean(overlap_mask_data)
        overlap_mask_data[overlap_mask_data >= overlap_mask_mean] = True
        overlap_mask_data[overlap_mask_data < overlap_mask_mean] = False
        overlap_mask_data[np.isnan(overlap_mask_data)] = False

        if overlap == "overlap":
            for [idx1, top_array1], top_array2 in zip(enumerate(mask_data), overlap_mask_data):
                for [idx11, sub_array1], sub_array2 in zip(enumerate(top_array1), top_array2):
                    for [idx111, element1], element2 in zip(enumerate(sub_array1), sub_array2):
                        if element1 == True and element1 == element2:
                            mask_data[idx1, idx11, idx111] = True
                        else:
                            mask_data[idx1, idx11, idx111] = False
            result_data = mask_data

        elif overlap == "no overlap":
            for [idx1, top_array1], top_array2 in zip(enumerate(mask_data), overlap_mask_data):
                for [idx11, sub_array1], sub_array2 in zip(enumerate(top_array1), top_array2):
                    for [idx111, element1], element2 in zip(enumerate(sub_array1), sub_array2):
                        if element1 == element2:
                            mask_data[idx1, idx11, idx111] = False
            result_data = mask_data

        elif overlap == "combine":
            for [idx1, top_array1], top_array2 in zip(enumerate(mask_data), overlap_mask_data):
                for [idx11, sub_array1], sub_array2 in zip(enumerate(top_array1), top_array2):
                    for [idx111, element1], element2 in zip(enumerate(sub_array1), sub_array2):
                        if element1 or element2:
                            mask_data[idx1, idx11, idx111] = True
            result_data = mask_data
        
        if dilation:
            result_data = sp.ndimage.binary_dilation(result_data)
            
        new_mask = nil.image.new_img_like(new_mask, result_data, affine=ref_affine)
        
        if dilation or side is not None:
            new_mask_p = "{}_".format(ori_mask[:-7])
            
            if dilation:
                new_mask_p = new_mask_p + "_dilation"
            
            if side is not None:
                new_mask_p = new_mask_p + "_{}".format(side)
                
            new_mask_p = new_mask_p + "{}_overlap.nii.gz".format(overlap_mask_p.split("/")[-1].split("_", 1)[0])
        else:
            new_mask_p = "{}_{}.nii.gz".format(ori_mask[:-7], overlap_mask_p.split("/")[-1].split("_", 1)[0])
        
        nib.save(new_mask, new_mask_p)
    return new_mask_p
```
